# **1. Cifrado del Mensaje**

## **1.1 Cifrado César**

```python
mensaje = input("Ingrese el Mensaje").lower()
clave = int(input("Ingrese la clave"))

alfabeto = list(string.ascii_lowercase)

cesar =  alfabeto[:clave]
cesar = alfabeto[clave:]+cesar

cesar_dict = dict(zip(alfabeto,cesar))

print("El mensaje ingresado fue: ", mensaje)
print("La clave ingresada fue: ", clave)
```

```python
palabra_cifrada_cesar = ""

for letra in mensaje:
    palabra_cifrada_cesar += cesar_dict[letra]

print("El cifrado de la palabra es: ",palabra_cifrada_cesar)
```

## **1.2 Cifrado Vigénere**

```python
mensaje_vig = input("Ingrese el mensaje").lower()
clave_vig = input("Ingrese la clave").lower()
```

```python
vigerene_dict = {}

for i in range(len(alfabeto)):
    for j in range(len(alfabeto)):
        a,b = alfabeto[i],alfabeto[j]
        index = (i+j) % 26
        vigerene_dict[(a,b)] = alfabeto[index]

print(vigerene_dict)
```

```python
cifrado_vigerene = ""

for letra_mensaje, letra_clave in zip(mensaje_vig,clave_vig):
    cifrado_vigerene+=vigerene_dict[(letra_mensaje,letra_clave)]

print("La palabra cifrada con el vigérene es: ", cifrado_vigerene)
```
